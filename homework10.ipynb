```python
#lets make a deal 3 doors
#No user input, random function acts as player 
import numpy as np
from numpy import random

wins_ns= 0
wins_s= 0
switched = 0
lost= 0
stayed = 0

wins_spoints = []
wins_nspoints = []
switchedpoints = []
stayedpoints = []
lost_spoints = []

for x in range(0,1001):
    
    list1 = ['a','b','c']
    door = random.choice(list1) # door player chooses
    #print(door)
    prize = random.choice(list1) #door with prize randomly chosen
    #print(prize)

    if (((prize == 'a')| (prize =='b')) & ((door == 'a') | (door == 'b'))):
        list1.remove('c') #game maker opens door c/3
        #print(list1)
    elif (((prize == 'b')| (prize == 'c')) & ((door == 'b') | (door == 'c'))):
        list1.remove('a') #game maker opens door a/1
        #print(list1)
    elif (((prize == 'c')| (prize == 'a')) & ((door == 'c') | (door == 'a'))):
        list1.remove('b') #game maker opens door b/2
        #print(list1)
    else:
        print("wrong")

    list2 = ['y','n']
    switch = random.choice(list2)
    #print(switch)

    points = 0
    if (((prize) == (door)) & (switch == 'y')):
        points += 1
        lost += points
        switched += points
    elif (((prize) == (door)) & (switch == 'n')):
        points +=1
        wins_ns += points
        stayed += points
    elif (((prize) != (door)) & (switch == 'y')):
        points +=1
        wins_s += points
        switched += points
    elif (((prize) != (door)) & (switch == 'n')):
        points +=1
        lost += points
        stayed += points
    else:
        print("error")

    wins_spoints.append(wins_s)
    wins_nspoints.append(wins_ns)
    switchedpoints.append(switched)
    stayedpoints.append(stayed)
    lost_spoints.append(lost)

print("Percentage winning when you switched",((wins_spoints[-1])/(switchedpoints[-1]))*100) #multiplied by 100 to get into percentage
print("Percentage winning when you did not switch",((wins_nspoints[-1])/(stayedpoints[-1]))*100)
```
